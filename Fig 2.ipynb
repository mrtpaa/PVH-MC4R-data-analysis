```python
#-----------Fig 2a----------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
import pprint

# Initialize dictionaries and lists
mouse_vectors = {'NCD': {}, 'HFD': {}}
diets = {}

mean_dict = {}
sem_dict = {}
smoothed_mean_vectors = {}
smoothed_sem_vectors = {}

# Iterate through the rows of the DataFrame
colors = {'NCD': 'navy', 'HFD': 'crimson'}
    
all_photometry_data = {}
original_shapes = {}  # Store original shapes before concatenation
triggered_columns_list = {}
padded_matrices = {}

rolling_window_size = 5

# Filter dataframe for the current mouse


# Process data
for r, row in df.iterrows():

    diet = row.diet
    mouse = row.mouse
    # Use mouse from row, but check if it's in mice_list

    if row.state != 'bad':

        # Determine photometry directory based on rig setup
        photometrydir = {
            'rbg': photometrydir_rbg,
            'roger': photometrydir_roger,
            'ch3rbg': photometrydir_ch3rbg,
            'ch3roger': photometrydir_ch3roger
        }.get(row.rig, None)
        if photometrydir is None:
            continue  # Skip if no valid directory

        # Locate the required files
        ensure_arr = glob.glob(fr'{photometrydir}/{row.mouse}/{row.date}_{row.mouse}/*-001_signal_ensuremat_all.csv')
        fiber_arr = glob.glob(fr'{photometrydir}/{row.mouse}/{row.date}_{row.mouse}/*-001_signal_trigmat_norm_df_new.csv')

        for file_path in ensure_arr:
            ensure = pd.read_csv(file_path, index_col=None, header=None)
            if ensure.shape[1] < 90:
                cols_to_add = 90 - ensure.shape[1]
                nan_columns = pd.DataFrame(np.nan, index=ensure.index, columns=range(ensure.shape[1], ensure.shape[1] + cols_to_add))
                ensure = pd.concat([ensure, nan_columns], axis=1)
            triggered = ensure[ensure.columns[ensure.max(axis=0) > 2]]

            if triggered.shape[1] > 29:
                triggered_names = triggered.columns.values

                # Initialize nested dictionaries if not already present
                if diet not in all_photometry_data:
                    all_photometry_data[diet] = {}
                    original_shapes[diet] = {}
                    triggered_columns_list[diet] = {}

                if mouse not in all_photometry_data[diet]:
                    all_photometry_data[diet][mouse] = []
                    original_shapes[diet][mouse] = []
                    triggered_columns_list[diet][mouse] = []

                # Get fiber files
                
                for file_path2 in fiber_arr:
                    photometry = pd.read_csv(file_path2, index_col=None, header=None)
                    photometry = photometry.rolling(window=10, min_periods=1).mean()

                    if photometry.shape[1] < 90:
                        cols_to_add = 90 - photometry.shape[1]
                        nan_columns = pd.DataFrame(np.nan, index=photometry.index, columns=range(photometry.shape[1], photometry.shape[1] + cols_to_add))
                        photometry = pd.concat([photometry, nan_columns], axis=1)

                    original_shapes[diet][mouse].append(photometry.shape)  # Store shape
                    triggered_columns_list[diet][mouse].append(triggered_names)  # Store triggered columns for this file
                    all_photometry_data[diet][mouse].append(photometry.values) 


  # **Concatenate and process photometry data**
for diet, mice_data in all_photometry_data.items():
    for mouse, photometry_list in mice_data.items():
        if photometry_list:
            photometry_concatenated = np.concatenate(photometry_list, axis=0)
            zscored_photometry = st.zscore(photometry_concatenated, axis=0, nan_policy='omit')

            print(f"\nZ-scored Data for {mouse}:")
            print(zscored_photometry)

            # **Split back to original shapes**
            index = 0
            for i, (rows, cols) in enumerate(original_shapes[diet][mouse]):
                zscored_photometry_segment = zscored_photometry[index:index + rows, :cols]
                index += rows  # Move index forward

                photometry_zscored = pd.DataFrame(zscored_photometry_segment, columns=range(cols))
                print(photometry_zscored.shape)

                # Retrieve the correct triggered column names for this recording
                triggered_names2 = triggered_columns_list[diet][mouse][i]
                print(triggered_names2)

                photometry_trig = photometry_zscored[triggered_names2]
                print(photometry_trig.shape)
                
                if photometry_trig.shape[1] > 29:
                    
#                             if photometry_trig.shape[1] > 29: 
#                                # photometry_trig.columns = range(len(triggered_cols.columns))
#                                 valid_columns = [col for col in triggered_names if col in photometry_trig.columns]

#                                 if valid_columns:  # Ensure there are valid columns to process
#                                     photometry_trig = photometry_trig[valid_columns]
#                                     photometry_trig.columns = range(photometry_trig.shape[1])  # Reindex columns

                    # Calculate the mean vector for the desired range
                    #photometry_trig = photometry_trig - photometry_trig.iloc[:500].mean()
        
                    mean_vector = photometry_trig.iloc[550:1000, :].mean(axis=0).to_numpy()

                    print(len(mean_vector))
                    # Calculate averages for every 5 columns and store
                    # mean_vector_avg_5 = pd.Series(mean_vector).rolling(window=5, min_periods=1, center=True).mean().to_numpy()
                    mean_vector_avg_5 = [
                        np.nanmean(mean_vector[i:i + 5]) for i in range(0, len(mean_vector), 5)]

                    if diet not in mouse_vectors:
                        mouse_vectors[diet] = {}
                    if mouse not in mouse_vectors[diet]:
                        mouse_vectors[diet][mouse] = []
                    mouse_vectors[diet][mouse].append(mean_vector_avg_5)
                    
                    #Export csv files-------
                    os.makedirs(save_path2, exist_ok=True)
                    filename = f"{diet}_{mouse}_response_to_ensure_{len(mouse_vectors[diet][mouse])}.csv"
                    file_path = os.path.join(save_path2, filename)

                    pd.DataFrame([mean_vector_avg_5]).to_csv(file_path, index=False, header=False)
                    #----End csv file export
                    
                    if mouse not in diets:
                        diets[mouse] = []
                    diets[mouse].append(diet)          


# Ensure padded_vectors dictionary is initialized
padded_vectors = {'NCD': {}, 'HFD': {}}

max_length = {'NCD': 0, 'HFD': 0}

mouse_counts = {}
for diet, mice in mouse_vectors.items():
    mouse_counts[diet] = len(mice)

print("Number of mice per condition/diet:")
pprint.pprint(mouse_counts)


# Determine max length per diet
for diet, mouse_entries in mouse_vectors.items():
    for mouse, vectors in mouse_vectors[diet].items():
        for vector in vectors:
            max_length[diet] = max(max_length[diet], len(vector))

# Pad vectors
for diet, mouse_entries in mouse_vectors.items():
    for mouse, vectors in mouse_entries.items():
        
        if diet not in padded_vectors:
            padded_vectors[diet] = {}
        if mouse not in padded_vectors[diet]:
            padded_vectors[diet][mouse] = []
        
        for vector in vectors:
            padded_vector = np.full(max_length[diet], np.nan)
            padded_vector[:len(vector)] = vector
            padded_vectors[diet][mouse].append(padded_vector)

# Compute means and SEMs
mean_dict = {'NCD': [], 'HFD': []}
sem_dict = {'NCD': [], 'HFD': []}

for diet, mouse_entries in padded_vectors.items():
    mouse_means_all = []  # Store all mouse means
    
    for mouse, vectors in mouse_entries.items():
        meal_data = np.array(vectors)  # Convert to NumPy array

        # Compute mean per mouse
        data_mean = np.nanmean(meal_data, axis=0)  # Ensure 1D
        
        mouse_means_all.append(data_mean)

    # Convert to array and calculate mean & SEM
    mouse_means_all = np.array(mouse_means_all)
    mean_dict[diet] = np.nanmean(mouse_means_all, axis=0)
    sem_dict[diet] = np.nanstd(mouse_means_all, axis=0) / np.sqrt(mouse_means_all.shape[0])
                  
rolling_window_size = 5  # Set the rolling window size

for diet in ['NCD', 'HFD']:
    if diet in mean_dict:
        # Apply rolling mean to the mean vector
        smoothed_mean_vectors[diet] = (
            pd.Series(mean_dict[diet])
            .rolling(window=rolling_window_size, min_periods=1, center=True)
            .mean()
            .to_numpy()
        )

        # Apply rolling mean to the SEM vector for consistency
        smoothed_sem_vectors[diet] = (
            pd.Series(sem_dict[diet])
            .rolling(window=rolling_window_size, min_periods=1, center=True)
            .mean()
            .to_numpy()
        )

# Plot the smoothed data
plt.figure(figsize=(3.85, 4))
xticks = np.linspace(0, 18, 10)
yticks = np.linspace(0, 2, 9)
colors = {'NCD': 'navy', 'HFD': 'crimson'}

# Plot settings
plt.xlabel("Trial", fontsize=18)
plt.ylabel("Response to Ensure\n[Z-score]", fontsize=18)
plt.title('Mean norm_df_3 zscored', fontsize=18, pad =20)
plt.ylim([0, 2])
plt.xlim([0, len(smoothed_mean_vectors['NCD'])])
plt.xticks(xticks, labels=['0', '', '20', '', '40', '', '60', '', '80', ''], fontsize=18)
plt.yticks(yticks, labels=['0', '', '0.5', '', '1', '', '1.5', '', '2'], fontsize=18)

# Plot each diet with SEM shading
for diet in ['NCD', 'HFD']:
    if diet in smoothed_mean_vectors:
        color = colors[diet]
        mean_vec = smoothed_mean_vectors[diet]
        sem_vec = smoothed_sem_vectors[diet]
        x = np.arange(len(mean_vec))

        plt.plot(x, mean_vec, color=color, label=f'{diet}')
        plt.fill_between(x, mean_vec - sem_vec, mean_vec + sem_vec, alpha=0.5, edgecolor=color, facecolor=color)

plt.legend(loc='lower right', fontsize=15)

# Save the plot
plt.tight_layout()
file_name2 = "Fig 2a.png"
full_save_path = os.path.join(save_path, file_name2)
plt.savefig(full_save_path, bbox_inches='tight')
plt.show()
```
