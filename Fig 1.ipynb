```python
#------MOUSE MEANS for NCD or HFD mice
#-------change the data folder to pick either directory



import pandas as pd
import numpy as np
import seaborn as sns
import glob
import os
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.gridspec import GridSpec


N = 5 # Aggregation factor for licking data
second_N = 10

mean_vectors = []
mean_licks = []

all_photometry_data = {}
original_shapes = {}  # Store original shapes before concatenation
triggered_columns_list = {}
padded_matrices = {}

mouse_data_dict = {
    'NCD': {},
    'HFD': {}}

for r, row in df.iterrows():
    print(row.mouse, row.date, row.photom)

    diet = row.diet.strip().upper()
    mouse = row.mouse


    # Set photometry directory based on diet. comment out as necessary
    if diet == 'NCD':
        photometrydir = photometrydir_ncd 
    #if diet == 'HFD':
        #photometrydir = photometrydir_hfd

        # Locate the required files
        ensure_arr = glob.glob(fr'{photometrydir}/{row.mouse}/{row.date}_{row.mouse}/*-001_signal_ensuremat_all.csv')
        licking_arr = glob.glob(fr'{photometrydir}/{row.mouse}/{row.date}_{row.mouse}/*-001_signal_lickmat.csv')
        running_arr = glob.glob(fr'{photometrydir}/{row.mouse}/{row.date}_{row.mouse}/*-001_signal_speedmat.csv')
        fiber_arr = glob.glob(fr'{photometrydir}/{row.mouse}/{row.date}_{row.mouse}/*trigmat_raw_norm_df_3.csv')

        for file_path in ensure_arr:
            ensure = pd.read_csv(file_path, index_col=None, header=None)
            triggered = ensure[ensure.columns[ensure.max(axis=0) > 2]]
            #extract only triggered trials
            if triggered.shape[1] > 29:  # Check if the number of columns is greater than 30
            
                triggered_names = triggered.columns.values
            
                    # Initialize nested dictionaries if not already present
                if diet not in all_photometry_data:
                    all_photometry_data[diet] = {}
                    original_shapes[diet] = {}
                    triggered_columns_list[diet] = {}

                if mouse not in all_photometry_data[diet]:
                    all_photometry_data[diet][mouse] = []
                    original_shapes[diet][mouse] = []
                    triggered_columns_list[diet][mouse] = []

                for file_path2 in fiber_arr:
                    photometry = pd.read_csv(file_path2, index_col=None, header=None)
                    #photometry = photometry.iloc[:3000,:]

                    if photometry.shape[1] < 60:
                        cols_to_add = 60 - photometry.shape[1]
                        nan_columns = pd.DataFrame(np.nan, index=photometry.index, columns=range(photometry.shape[1], photometry.shape[1] + cols_to_add))
                        photometry = pd.concat([photometry, nan_columns], axis=1)
                        
                    original_shapes[diet][mouse].append(photometry.shape)  # Store shape
                    triggered_columns_list[diet][mouse].append(triggered_names)  # Store triggered columns for this file
                    all_photometry_data[diet][mouse].append(photometry.values) 
                    print(f"\n All photometry data for {mouse}:")

                for file_path3 in licking_arr:
                    licking = pd.read_csv(file_path3, index_col=None, header=None)
                    licking = licking[triggered_names]
                    if licking.shape[1] > 29:

                        licking_aggregated_agg = licking.groupby(licking.index // N).sum().rolling(window=second_N, min_periods=1).mean()
                        mean_licking = (licking_aggregated_agg.iloc[:, 0:60].mean(axis=1) * 10).to_numpy()
                        mean_licks.append(mean_licking)
                        
                        print(f"\n Licking for {mouse}:")
                        #print(mean_licks)
                        
                        # --- CSV EXPORT: Mean licking
                        licking_filename = os.path.join(save_path2, f"mean_licking_{mouse}_{len(mean_licks)}.csv")
                        pd.DataFrame(mean_licking).to_csv(licking_filename, index=False, header=False)
                        print(f"Exported licking data to {licking_filename}")
                        
                for file_path4 in running_arr:
                    df_signal = pd.read_csv(file_path4, index_col=None, header=None)
                    df_signal = df_signal[triggered_names]
                    if df_signal.shape[1] > 29:
                        #print(df_signal)
                        # Aggregate the data by taking the mean of every N rows
                        df_aggregated = df_signal.groupby(df_signal.index // N).mean()
                        # Calculate the mean of each row for the first 60 columns
                        mean_vector = df_aggregated.iloc[:, 0:60].mean(axis=1).to_numpy()
                        mean_vectors.append(mean_vector)
                        print(diet)
                        print(f"\n Running for {mouse}:")
                       
                    
                        # --- CSV EXPORT: Mean running vector
                        running_filename = os.path.join(save_path2, f"mean_running_{mouse}_{len(mean_vectors)}.csv")
                        pd.DataFrame(mean_vector).to_csv(running_filename, index=False, header=False)
                        print(f"Exported running vector to {running_filename}")
                        

# **Concatenate and process photometry data**
for diet, mice_data in all_photometry_data.items():
    for mouse, photometry_list in mice_data.items():
        if photometry_list:
            photometry_concatenated = np.concatenate(photometry_list, axis=0)
            zscored_photometry = st.zscore(photometry_concatenated, axis=0, nan_policy='omit')

            print(f"\nZ-scored Data for {mouse}:")
            print(zscored_photometry)

            # **Split back to original shapes**
            index = 0
            for i, (rows, cols) in enumerate(original_shapes[diet][mouse]):
                zscored_photometry_segment = zscored_photometry[index:index + rows, :cols]
                index += rows  # Move index forward

                photometry_zscored = pd.DataFrame(zscored_photometry_segment, columns=range(cols))
                print(photometry_zscored.shape)

                # Retrieve the correct triggered column names for this recording
                triggered_names2 = triggered_columns_list[diet][mouse][i]
                print(triggered_names2)

                photometry_trig = photometry_zscored[triggered_names2]
                print(photometry_trig.shape)

                if photometry_trig.shape[1] > 29:
                    data = photometry_trig.groupby(photometry.index // N).mean()
                    data = data.rolling(window=5, min_periods=1).mean()
                    print(data.shape)
                    
                    matrix = data.values
                    #print(matrix)
                    
                    print(diet)
                    print(mouse)
                    if diet not in mouse_data_dict:
                        mouse_data_dict[diet] = {}
                    if mouse not in mouse_data_dict[diet]:
                        mouse_data_dict[diet][mouse] = []

                    # Append data correctly
                    mouse_data_dict[diet][mouse].append(matrix)
                    
                    # --- CSV EXPORT: Matrix
                    matrix_filename = os.path.join(save_path2, f"matrix_{mouse}_{len(mouse_data_dict[diet][mouse])}.csv")
                    pd.DataFrame(matrix).to_csv(matrix_filename, index=False, header=False)
                    print(f"Exported matrix to {matrix_filename}")


                    
import pprint
pprint.pprint(mouse_data_dict)

if len(mouse_data_dict) > 0:
    # Determine the maximum number of columns among all matrices
    max_columns = max(matrix.shape[1] for diet in mouse_data_dict.values() for matrix_list in diet.values() for matrix in matrix_list)
    
    for diet, mouse_entries in mouse_data_dict.items():
        print(f"Checking diet: {diet}, type: {type(mouse_entries)}")
        for mouse, matrix_list in mouse_entries.items():
            for matrix in matrix_list:
                if matrix.shape[1] < 60:
                    padding = np.full((matrix.shape[0], 60 - matrix.shape[1]), np.nan)
                    padded_matrix = np.hstack((matrix, padding))
                else:
                    padded_matrix = matrix
                    
                if diet not in padded_matrices:
                    padded_matrices[diet] = {}
                if mouse not in padded_matrices[diet]:
                    padded_matrices[diet][mouse] = []

                # Append the padded matrix
                padded_matrices[diet][mouse].append(padded_matrix)            
               
        
for diet, mouse_entries in padded_matrices.items():
    mouse_means_early = []  # Store all mouse means for Early Meal

    for mouse, values in mouse_entries.items():
        # Convert padded matrices to a 3D array and calculate the mean
        stacked_matrices = np.stack(values)
        mean_across_matrices = np.nanmean(stacked_matrices, axis=0)  # Ignore NaNs in mean calculation
        df_mean_across_matrices = pd.DataFrame(mean_across_matrices)
        print(df_mean_across_matrices.shape)

df_mean_across_matrices = df_mean_across_matrices - df_mean_across_matrices.iloc[:100].mean()        

# Calculate mean and SEM for mean vectors running
mean_vectors_df = pd.DataFrame(mean_vectors)
mean_df = np.mean(mean_vectors_df, axis=0)
sem_df = np.std(mean_vectors_df, axis=0) / np.sqrt(mean_vectors_df.shape[0])

# Calculate mean and SEM for mean vectors running
mean_licks_df = pd.DataFrame(mean_licks)
mean_licks_df = np.mean(mean_licks_df, axis=0)
sem_licks_df = np.std(mean_licks_df, axis=0) / np.sqrt(mean_licks_df.shape[0])

# Plot combined figure
fig = plt.figure(figsize=(7, 5))
gs = GridSpec(2, 2, width_ratios=[20, 1], height_ratios=[0.5, 1.5], hspace=0.1, wspace=0.05)

ax1 = fig.add_subplot(gs[0, 0])
x_df = np.arange(len(mean_df))
ax1.plot(x_df, mean_df, color='navy', linewidth=1.2)
ax1.fill_between(x_df, mean_df - sem_df, mean_df + sem_df, alpha=0.3, edgecolor='navy', facecolor='navy')
ax1.set_ylim(0, 8)
ax1.set_yticklabels(ax1.get_yticks(), fontsize=15)
ax1.set_xticks([])
ax1.axvline(x=100, color='orange', linewidth=1.2, alpha=0.8)
ax1.axvline(x=110, color='purple', linewidth=1.2, alpha=0.8)
ax1.set_ylabel("Running\nspeed\n[cm/s]", fontsize=15, color='navy')
ax1.tick_params(axis='x', labelbottom=False)

ax3 = ax1.twinx()
x_licks_df = np.arange(len(mean_licks_df))
ax3.plot(x_licks_df, mean_licks_df, color='crimson', linewidth=1.2)
ax3.fill_between(x_licks_df, mean_licks_df - sem_licks_df, mean_licks_df + sem_licks_df, alpha=0.3, edgecolor='crimson', facecolor='crimson')
ax3.set_ylim(0, 8)
ax3.set_yticklabels(ax3.get_yticks(), fontsize=15)
ax3.set_ylabel("Lick rate\n[licks/s]", fontsize=15, color='crimson')

# Plot heatmap
ax2 = fig.add_subplot(gs[1, 0], sharex=ax1)
sns.heatmap( df_mean_across_matrices.T, ax=ax2, cmap="vlag", yticklabels=False, vmin=-2.5, vmax=2.5, cbar=False)

ax2.set_xlabel("Time [s]", fontsize=15)
ax2.set_ylabel("Trials with Ensure", fontsize=15)
xticks2 = np.linspace(0, 600, 7)
ax2.xaxis.set_ticks(xticks2)
ax2.set_xticklabels(['-10', '0', '10', '20', '30', '40', '50'], fontsize=15)
ax2.set_yticks(np.linspace(0, 60, 7))
ax2.set_yticklabels(['0', '10', '20', '30', '40' ,'50', '60'], fontsize=15)
ax2.axvline(x=100, color='orange', linewidth=1.2, alpha=0.8, label='Tone')
ax2.axvline(x=110, color='purple', linewidth=1.2, alpha=0.8, label='Ensure')
ax2.legend(loc='upper right', fontsize = 15)

#Add colorbar separately
cbar_ax = fig.add_subplot(gs[1, 1])  # Separate axis for the colorbar
norm = plt.Normalize(vmin=-2.5, vmax=2.5)
sm = plt.cm.ScalarMappable(cmap="vlag", norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, cax=cbar_ax, orientation='vertical', pad = 0.01 )
cbar_ax.set_aspect(0.5)
cbar.set_label('Response to Ensure [Z-score]', fontsize=15)  # Set colorbar label font size
cbar.ax.tick_params(labelsize = 15)

# Save the figure
heatmap_file_name = f"NCD_Fig1 ef.png"
heatmap_save_path = os.path.join(save_path, heatmap_file_name)
plt.savefig(heatmap_save_path, bbox_inches='tight')
print(f"Combined plot saved for.")
plt.show()


```
